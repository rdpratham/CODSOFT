```python
import pandas as pd
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.Class.value_counts()
```

```python
legit = df[df['Class']==0]
fraud = df[df['Class']==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
legit_s = legit.sample(n=492)
```

```python
legit_s.shape
```

```python
fraud.shape
```

```python
ndf = pd.concat([fraud,legit_s],axis=0)
```

```python
ndf.head()
```

```python
ndf.Class.value_counts()
```

```python
x=ndf.drop(columns='Class',axis=1)
```

```python
y = ndf['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test  = train_test_split(x,y,stratify=y,random_state=2529)
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
md = LogisticRegression(max_iter=500)
```

```python
md.fit(x_train,y_train)
```

```python
y_p = md.predict(x_test)
```

```python
md.coef_
```

```python
md.intercept_
```

```python
from sklearn.metrics import accuracy_score,r2_score
```

```python
accuracy_score(y_p,y_test)
```

```python
r2_score(y_p,y_test)
```
